# 1. LASER Diode Characteristics
#### Aim: To study the V-I characteristics of semiconductor diode laser.
#### Apparatus: Laser diode, Photo Detector, 0-5V de power supply, Voltmeter (0-20 V), Milli ammeter (0-20 mA).
#### Circuit diagram:
![SNOWFALL](Img/EXP1.jpg) 

#### Formula: 
#### 入＝ hc / Eg ＝ 12400 / Eg Å
#### Where,  入 - Wavelength, h - Plank's constant, Eg - Energy gap, c - Speed of light

```python
import matplotlib.pyplot as plt
```

```python
voltage_input = input("Enter Voltage(V): (seperated with ','): ")
voltage = [float(value) for value in voltage_input.split(',')]

c1_input = input("Enter Current(mA) - R 100Ω: (seperated with ','): ")
current_1 = [float(value) for value in c1_input.split(',')]

c2_input = input("Enter Current(mA) - R 150Ω: (seperated with ','): ")
current_2 = [float(value) for value in c2_input.split(',')]

c3_input = input("Enter Current(mA) - R 220Ω: (seperated with ','): ")
current_3 = [float(value) for value in c3_input.split(',')]
```

```python
# Create the V-I graph
plt.plot(voltage, current_1, marker='o', label='R = 100Ω')
plt.plot(voltage, current_2, marker='o', label='R = 150Ω')
plt.plot(voltage, current_3, marker='o', label='R = 220Ω')

# Add labels and title
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')
plt.title('V-I Charateristics Graph of LASER Diode')

# Add a legend
plt.legend()

# Show the graph
plt.show()
```
